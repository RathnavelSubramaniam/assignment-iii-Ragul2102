```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from ISLP import load_data
from ISLP.models import ModelSpec as MS
from ISLP import confusion_table
from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import \
 (LinearDiscriminantAnalysis as LDA,
 QuadraticDiscriminantAnalysis as QDA)
from sklearn.neighbors import KNeighborsClassifier
```

```python
Auto_data=load_data('Auto')
Auto_data
```

1.A

```python
median = Auto_data['mpg'].median()
median
```

```python
Auto_data['mpg01'] = (Auto_data['mpg'] > median).astype(int)
```

```python
Auto_data
```

1.B

```python
sns.scatterplot(x="mpg01", y="horsepower", data=Auto_data)
plt.title('Horsepower vs mpg01')
plt.show()
```

```python
sns.boxplot(x="mpg01", y="horsepower", data=Auto_data)
plt.title('Horsepower vs mpg01')
plt.show()
```

1.C

```python
x= Auto_data[['cylinders','displacement','horsepower']]
y= Auto_data['mpg01']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
Auto_data
```

D in B

#### Answer:

```python
lda = LDA(store_covariance=True)
```

```python
lda.fit(x_train, y_train)
```

```python
lda_pred = lda.predict(x_test)
```

```python
confusion_table(lda_pred, y_test)
```

E in B

```python
qda = QDA(store_covariance=True)
```

```python
qda.fit(x_train, y_train)
```

```python
qda_pred = qda.predict(x_test)
```

```python
confusion_table(qda_pred, y_test)
```

F in B

```python
x = MS(x_train).fit_transform(x_train)
y = y_train
result = sm.GLM(y, x, family=sm.families.Binomial()).fit()
print(result.summary()) 
```

G in B

```python
knn1 = KNeighborsClassifier(n_neighbors=1)
knn1.fit(x_train, y_train)
knn1_pred = knn1.predict(x_test)
confusion_table(knn1_pred, y_test)
```

```python
knn2 = KNeighborsClassifier(n_neighbors=2)
knn2.fit(x_train, y_train)
knn2_pred = knn2.predict(x_test)
confusion_table(knn2_pred, y_test)
```

```python
knn3 = KNeighborsClassifier(n_neighbors=3)
knn3.fit(x_train, y_train)
knn3_pred = knn3.predict(x_test)
confusion_table(knn3_pred, y_test)
```
